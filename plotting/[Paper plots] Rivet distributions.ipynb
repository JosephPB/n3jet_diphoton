```python
from rivet_plotter import RivetPlotter
```

```python
save_dir = './paper_plots/6_point/distributions/'
```

# Angular plots

## $d\phi_{jj}$

```python
dphi_jj_file = rivet_dir + 'dphijj.dat'
```

```python
dphi_jj_plotter = RivetPlotter(dphi_jj_file)
```

```python
dphi_jj_fig = dphi_jj_plotter.plot(
    xlabel = '$\Delta\phi_{jj}$ [rad]',
    ylabel = 'd$\sigma/$d$\Delta\phi_{jj}$ [fb rad$^{-1}$]'
)
```

```python
dphi_jj_fig.savefig(save_dir + 'dphi_jj.png', dpi=150, bbox_inches='tight')
```

## $R_{j_1,\gamma_1}$
